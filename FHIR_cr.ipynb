```python
from selenium import webdriver
from bs4 import BeautifulSoup,NavigableString,Tag
import requests

res=requests.get("http://hl7.org/fhir/resourcelist.html")
soup=BeautifulSoup(res.text)

temp_links=[]
sult_links=[]

for k in soup.find('div',{'id':'tabs'}).find_all('a'):
    try:
        temp_links.append("http://hl7.org/fhir/" +  k['href'])
    except:
        print("")
        
for x in temp_links:
    if "#" not in x:
        sult_links.append(x)
        sult_links2=sult_links[0:146]

print(sult_links2)
```
